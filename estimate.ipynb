```python
def mse_solver(X, Y):
    return (X*Y).sum() / (X*X).sum()
```

$$L = \sum_{i=0}^{N}{(f_i - ks_i)^2}$$

$$\frac{dL}{dk} = -\sum_{i=0}^{N}{2(f_i - ks_i)s_i} = 0$$

$$\sum_{i=0}^{N}{f_is_i - k{s_i}^2} = 0$$

$$\sum_{i=0}^{N}{f_is_i} = k\sum_{i=0}^{N}{s_i^2}$$

$$k = \frac{\sum_{i=0}^{N}{f_is_i}}{\sum_{i=0}^{N}{s_i^2}}$$
